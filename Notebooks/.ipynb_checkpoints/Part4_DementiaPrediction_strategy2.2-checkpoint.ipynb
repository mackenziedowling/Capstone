# Capstone: Dementia Classification

Relatively little is known about what causes Dementia in older age, as no one test can determine the diagnosis of Dementia in a patient.  Doctors diagnose Alzheimer's and other types of Dementia based on a combination of factors. Lab tests and other health factors are typically measured to eliminate all other possibilities. A diagnosis for Dementia includes careful examination of medical history, a physical examination, laboratory tests, and the characteristic changes in thinking, day-to-day function, and behavior associated with each type.  Through this type of analysis, doctors can determine that a person has Dementia with a high level of certainty. However, determining if a person will have Dementia is left with much uncertainty as there is no solution to testing and screening for the onset of Dementia before any symptoms.

# Model Preprocessing

```python
def evaluation(mdl, X_train, y_train, X_test, y_test):
    results = {}
    train_aucc = mdl.score(X_train, y_train)
    test_aucc = mdl.score(X_test, y_test)
    y_pred_train = mdl.predict(X_train)
    y_pred_test = mdl.predict(X_test)
    
    print(f'Training auccuracy: {train_aucc}')
    print(f'Test auccuracy: {test_aucc}')
    print()
    
    tn, fp, fn, tp = confusion_matrix(y_test, mdl.predict(X_test)).ravel() 
    plot_confusion_matrix(mdl, X_test, y_test, cmap='Blues', values_format='d');
    
    print('Order of classes: ', mdl.classes_)
    confusion_df = pd.DataFrame(
                    data=confusion_matrix(y_test, y_pred_test),
                    index=[f'actual {target_class}' for target_class in mdl.classes_],
                    columns=[f'predicted {target_class}' for target_class in mdl.classes_])
    print('\nTest Confusion Matrix: \n', confusion_df)
    
    print()
    print("True Negatives: %s" % tn)
    print("False Positives: %s" % fp)  
    print("False Negatives: %s" % fn)
    print("True Positives: %s" % tp)
    #print('\nTest Confusion Matrix: \n', confusion_df)
    print(classification_report(y_test, y_pred_test))
    # 5. Test recall and precision (if no samples, set recall/precision to 0)
    results['test_recall'] = recall_score(y_test, y_pred_test, 
                                          pos_label=1, zero_division=0)
    results['test_precision'] = precision_score(y_test, y_pred_test,
                                                pos_label=1, zero_division=0)
```

### Missing Values Imputation:

```python
knn_imputer = KNNImputer(n_neighbors=2)
X_train = knn_imputer.fit_transform(X_train)
X_test = knn_imputer.fit_transform(X_test)
```

### Scaling
Since the ranges of max and min for the Lab result vary by test and do not contain the same type of measurement for each test the best option to nuetralize this in modeling is two instantiate and transform the X_test and X_train with a standardscaler from SK.Learn libary. 

```python
# Instantiate our StandardScaler.
ss = StandardScaler()

# Standardize X_train.
X_train = ss.fit_transform(X_train)

# Standardize X_test.
X_test = ss.transform(X_test)
```

# Under Sampling Test:

```python
####### RandomUnderSampler ############################################
rs = RandomUnderSampler()
X_train_rs, y_train_rs = rs.fit_resample(X_train, y_train)

#fitting random forest model on Random Under sampler data
rf_randSamp = RandomForestClassifier(n_estimators=100)
rf_randSamp.fit(X_train_rs, y_train_rs)

#saving pr auc score
precision_rs, recall_rs, _ = precision_recall_curve(y_test, rf_randSamp.predict_proba(X_test)[:,1])
auc_score_rs = auc(recall_rs, precision_rs)

############## TomekLinks #####################################
tl = TomekLinks()
X_train_tl, y_train_tl =tl.fit_resample(X_train, y_train)

#fitting random forrest classifier baseline model on tl

rf_tl = RandomForestClassifier(n_estimators=100)
rf_tl.fit(X_train_tl, y_train_tl)

#saving pr auc score
precision_tl, recall_tl, _ = precision_recall_curve(y_test, rf_tl.predict_proba(X_test)[:,1])
auc_score_tl = auc(recall_tl, precision_tl)


###### Neighborhood Cleaning Rule #####################################
ncr = NeighbourhoodCleaningRule()
X_train_ncr, y_train_ncr = ncr.fit_resample(X_train, y_train)

#fitting random forrest classifier baseline model on NCR
rf_ncr = RandomForestClassifier(n_estimators=100)
rf_ncr.fit(X_train_ncr, y_train_ncr)

#saving pr auc score
precision_ncr, recall_ncr, _ = precision_recall_curve(y_test, rf_ncr.predict_proba(X_test)[:,1])
auc_score_ncr = auc(recall_ncr, precision_ncr)

##### Cluster Centorids ##############################################
cc = ClusterCentroids()
X_train_cc, y_train_cc = cc.fit_resample(X_train, y_train)

#fiting random forest classifier
rf_cc = RandomForestClassifier(n_estimators=100)
rf_cc.fit(X_train_cc, y_train_cc)

#saving pr auc score
precision_cc, recall_cc, _ = precision_recall_curve(y_test, rf_cc.predict_proba(X_test)[:,1])
auc_score_cc = auc(recall_cc, precision_cc)


#printing each score for comparison
print(f'Random UnderSampler: PR AUC {auc_score_rs}')
print(f'TomekLinks: PR AUC {auc_score_tl}')
print(f'Neighborhood Cleaning Rule: PR AUC {auc_score_ncr}')
print(f'Cluster Centroid: PR AUC {auc_score_cc}')
      
```

```python
#Creating a dictionary to input the class split 
#in each transformed y_train

# Create empty dictionary
resampling_results = {}

#add in RUS sampler class split
resampling_results['RandomUnderSampler'] = y_train_tl.value_counts(normalize=True)

#add in TomekLinks class split
resampling_results['Tomeklinks'] = y_train_tl.value_counts(normalize=True)

#add in NCR class split
resampling_results['Neighborhood Cleaning Rule'] = y_train_ncr.value_counts(normalize=True)

#add in cc class split
resampling_results['Cluster Centroid'] = y_train_cc.value_counts(normalize=True)

#transform to df
resampling_results = pd.DataFrame(resampling_results)

resampling_results = resampling_results.transpose()

#resetting index for evaluation comparrrison column
#resampling_results.reset_index(level=0, inplace=True)

# adding in the PR AUC Score for comparrison
resampling_results['PR-AUC score'] = [auc_score_rs, auc_score_tl, auc_score_ncr, auc_score_cc]


resampling_results
```

### 1) Decision Tree:

The DecisionTree Model offers a few benefits to prediction modeling with out data. We do not have to sale our data because the assumptions around how the data is used and estimated, don't relay on scale. We also discovered in EDA that the data was not normally distributed and partially skewed. Since decision trees are nonparametric which means we do not have to makes any assumptions about our data or the how errors are distributed. On the flip side Decisions trees can suffer from high erros due to variance and regularizations techniques will need to be used to to avoid this. In addition as we discovered through EDA the data has a large case of unbalanced clases. There are sevel undersampling techniques we can implement in preprocessing or use the class_weight parameter to address this. 


As stated above DecisionTree model will require a couple transformation steps to our training in order to run properly. Below Imputation for missing values is being adresssed with SimpleImputer with a median strategy as in EDA this imputation was the most robust toward outliers in the dataset.  

```python
dt_baseline = DecisionTreeClassifier()
dt_baseline.fit(X_train_ncr, y_train_ncr)

train_auc = dt_baseline.score(X_train, y_train)
test_auc = dt_baseline.score(X_test, y_test)
print(f'Decision Tree train accuracy: {train_auc}')
print(f'Decision Tree Test accuracy: {test_auc}') 

dt_precision, dt_recall, _ = precision_recall_curve(y_test, dt_baseline.predict_proba(X_test)[:,1])
print(auc(dt_recall, dt_precision))

plot_precision_recall_curve(dt_baseline, X_test, y_test, name = 'Decision Tree Classifier')
plt.legend(loc="upper center");
plt.title('Decision Tree PR curve');
```

### 2) RandomForrestClassifier

Random Forest is chosen to test because of the benefits it offers to the current dataset and the more robust ability to counteract overfitting. In general Random Forest classification differs from bagging decision trees in only one way. It uses a modified tree learning algorithm that selects, at each split in the learning process a random subset of the features. The typically helps to reduce correlation in features and two the classes whereas decision tree will use many or all features causing overfitting.

**For the prediction of dementia on the MIMIC III dataset that has been cleaned and explored this model offers a few benefits:**<br>
1) RandomForest handles errors in imbalanced data which means that even though weve under sampled the data the class balance will not impact the performance. <br>
<br>
2) Handles missing values. Unlike decision tree the Random Forest model can handle missing values and dat in the model and still maintain good results.<br>
<br>
3) Less overfitting tendency than Decision Tree<br>
<br>
The drawback that should be kept in mind for this model is features need to have some prediction power. This could be the reason this model does not perform well on the dataset since the EDA analysis show title correlation with features to classes differences.

# Hypertuning Best Model

```python
xgb = XGBClassifier(learning_rate=0.02, n_estimators=600, objective='binary:logistic',
                    silent=True, nthread=1)

params = {
    'min_child_weight': [1, 5, 10],
        'gamma': [0.5, 1, 1.5, 2, 5],
        'subsample': [0.6, 0.8, 1.0],
        'colsample_bytree': [0.6, 0.8, 1.0],
        'max_depth': [3, 4, 5]
}

folds = 5
param_comb = 5

skf = StratifiedKFold(n_splits=folds, shuffle = True, random_state = 42)


random_search = RandomizedSearchCV(xgb,
                                   param_distributions=params,
                                   n_iter=param_comb,
                                   cv = skf.split(X_train_ncr,y_train_ncr),
                                   verbose=2,
                                   scoring='average_precision',
                                   n_jobs = 6,
                                   random_state=42)
```

```python
random_search.fit(X_train_ncr, y_train_ncr)
```
