# Classifying Mental Illness

Relatively little is known about what causes Dementia in older age, as no one test can determine the diagnosis of Dementia in a patient.  Doctors diagnose Alzheimer's and other types of Dementia based on a combination of factors. Lab tests and other health factors are typically measured to eliminate all other possibilities. A diagnosis for Dementia includes careful examination of medical history, a physical examination, laboratory tests, and the characteristic changes in thinking, day-to-day function, and behavior associated with each type.  Through this type of analysis, doctors can determine that a person has Dementia with a high level of certainty. However, determining if a person will have Dementia is left with much uncertainty as there is no solution to testing and screening for the onset of Dementia before any symptoms.

The focus on this project is on prediction classification for dementia diagnosis using electronic health records of ICU patients from the MIMIC III database. The ultimate goal is to understand the role of the various features in the model's ability to differentiate patients diagnosed with Dementia vs. those who do not have Dementia. The prior notebook explored the solution approach by taking a random undersampling of the majority class to predict dementia patients directly. 

The alternative solution strategy executed in this notebook takes a two-fold approach. The first fold explores the classification prediction of patients within the broader ccs diagnosis category, Mental Health. The second fold then assumes given the patient is within the ccs-group of Mental Health can the classifier distinguish patients diagnosed with Dementia from the Mental Health patients.  The ultimate goal is to understand the distinguishing factors that separate dementia patients from the dataset's available features. 

## Solution Approach 2: Part 1 Classifying Mental Health

 Several classification models were tested via statical modeling, such as Decision Tree, Random Forest classification, Extra Trees Classification, and XG Boost. These models were selected for their ability to perform well on classification problems. The pros and cons of these models are addressed within each model section.

### Model Evaluation:
Due to the class imbalance present in the dataset, accuracy is not used for model evaluation. Instead, the classification models' overall performance is measured using the area under the receiver operating characteristics curve (AUROC) specifically using percision and recall. This is known as the PR AUC score and can be visualized with the PR ROC curve. This is used for evaluation since the dataset is extremely unbalanced. Percision is directly influenced by class imbalance so the percision-recall curve will highlight differences between models for highly unbalanced dataset. The precision is measuring the rate of correct positive predictions out of all positive predictions while recall is measuring the rate of correct positive predicitons out of all possible postive predictions.

# UnderSampling Test

In the exploratory analysis the data showed a large class imbalance. A class imbalance in classification poses a challenge for predicitive modeling since most machine learning algrithms assume equal number of examples for each class. The implication of imbalance of classes on model performance is poored predictive performance specifically for the minority class. 

The different undersampling techniques viewed in the exploratory analysis are below and will be tested against a baseline RandomForest classification model to compare performance. 

![title](../Documents/images/UnderSampling_2d.png)

Centroid Cluster This method undersamples the majority class by replacing a cluster of majority samples This method finds the clusters of majority class with K-mean algorithms. Then it keeps the cluster centroids of the N clusters as the new majority samples

```python
####### RandomUnderSampler ############################################
rs = RandomUnderSampler()
X_train_rs, y_train_rs = rs.fit_resample(X_train, y_train)

#fitting random forest model on Random Under sampler data
rf_randSamp = RandomForestClassifier(n_estimators=100)
rf_randSamp.fit(X_train_rs, y_train_rs)

#saving pr auc score
precision_rs, recall_rs, _ = precision_recall_curve(y_test, rf_randSamp.predict_proba(X_test)[:,1])
auc_score_rs = auc(recall_rs, precision_rs)

############## TomekLinks #####################################
tl = TomekLinks()
X_train_tl, y_train_tl =tl.fit_resample(X_train, y_train)

#fitting random forrest classifier baseline model on tl


rf_tl = RandomForestClassifier(n_estimators=100)
rf_tl.fit(X_train_tl, y_train_tl)

#saving pr auc score
precision_tl, recall_tl, _ = precision_recall_curve(y_test, rf_tl.predict_proba(X_test)[:,1])
auc_score_tl = auc(recall_tl, precision_tl)


###### Neighborhood Cleaning Rule #####################################
ncr = NeighbourhoodCleaningRule()
X_train_ncr, y_train_ncr = ncr.fit_resample(X_train, y_train)

#fitting random forrest classifier baseline model on NCR
rf_ncr = RandomForestClassifier(n_estimators=100)
rf_ncr.fit(X_train_ncr, y_train_ncr)

#saving pr auc score
precision_ncr, recall_ncr, _ = precision_recall_curve(y_test, rf_ncr.predict_proba(X_test)[:,1])
auc_score_ncr = auc(recall_ncr, precision_ncr)

##### Cluster Centorids ##############################################
cc = ClusterCentroids()
X_train_cc, y_train_cc = cc.fit_resample(X_train, y_train)

#fiting random forest classifier
rf_cc = RandomForestClassifier(n_estimators=100)
rf_cc.fit(X_train_cc, y_train_cc)

#saving pr auc score
precision_cc, recall_cc, _ = precision_recall_curve(y_test, rf_cc.predict_proba(X_test)[:,1])
auc_score_cc = auc(recall_cc, precision_cc)


#printing each score for comparison
print(f'Random UnderSampler: PR AUC {auc_score_rs}')
print(f'TomekLinks: PR AUC {auc_score_tl}')
print(f'Neighborhood Cleaning Rule: PR AUC {auc_score_ncr}')
print(f'Cluster Centroid: PR AUC {auc_score_cc}')
      
```

```python
#Creating a dictionary to input the class split 
#in each transformed y_train

# Create empty dictionary
resampling_results = {}

#add in RUS sampler class split
resampling_results['RandomUnderSampler'] = y_train_tl.value_counts(normalize=True)

#add in TomekLinks class split
resampling_results['Tomeklinks'] = y_train_tl.value_counts(normalize=True)

#add in NCR class split
resampling_results['Neighborhood Cleaning Rule'] = y_train_ncr.value_counts(normalize=True)

#add in cc class split
resampling_results['Cluster Centroid'] = y_train_cc.value_counts(normalize=True)

#transform to df
resampling_results = pd.DataFrame(resampling_results)

resampling_results = resampling_results.transpose()

#resetting index for evaluation comparrrison column
#resampling_results.reset_index(level=0, inplace=True)

# adding in the PR AUC Score for comparrison
resampling_results['PR-AUC score'] = [auc_score_rs, auc_score_tl, auc_score_ncr, auc_score_cc]


resampling_results
```

**UnderSampler Method Interpretation:** 

The results from testing different imbalance class under-samplers vary. In terms of balancing the classes the **cluster centroid** method did the best by creating perfectly balanced classes split 50% by 50%. 

When evaluating how these under-samplers will perform with modeling the **PR AUC** score is a good measurement. Precision is directly influenced by class imbalance so the Precision-recall curves is informative measurement to highlight differences between models for highly imbalanced data sets. When you compare different models with imbalanced settings, the area under the Precision-Recall curve is more sensitive than say the area under the ROC curve. Based on this test **TomekLink under-sampler** performed the best with the highest **PR AUC score of 0.39%**. In the following model selection this under-sampler will be used for data preprocessing.

```python
tl = TomekLinks(sampling_strategy='majority')
X_train_z, y_train_z =tl.fit_resample(X_train, y_train)
```

```python
def evaluation(mdl, X_train, y_train, X_test, y_test):
    results = {}
    train_aucc = mdl.score(X_train, y_train)
    test_aucc = mdl.score(X_test, y_test)
    y_pred_train = mdl.predict(X_train)
    y_pred_test = mdl.predict(X_test)
    
    print(f'Training auccuracy: {train_aucc}')
    print(f'Test auccuracy: {test_aucc}')
    print()
    
    tn, fp, fn, tp = confusion_matrix(y_test, mdl.predict(X_test)).ravel() 
    plot_confusion_matrix(mdl, X_test, y_test, cmap='Blues', values_format='d');
    
    print('Order of classes: ', mdl.classes_)
    confusion_df = pd.DataFrame(
                    data=confusion_matrix(y_test, y_pred_test),
                    index=[f'actual {target_class}' for target_class in mdl.classes_],
                    columns=[f'predicted {target_class}' for target_class in mdl.classes_])
    print('\nTest Confusion Matrix: \n', confusion_df)
    
    print()
    print("True Negatives: %s" % tn)
    print("False Positives: %s" % fp)  
    print("False Negatives: %s" % fn)
    print("True Positives: %s" % tp)
    #print('\nTest Confusion Matrix: \n', confusion_df)
    print(classification_report(y_test, y_pred_test))
    # 5. Test recall and precision (if no samples, set recall/precision to 0)
    results['test_recall'] = recall_score(y_test, y_pred_test, 
                                          pos_label=1, zero_division=0)
    results['test_precision'] = precision_score(y_test, y_pred_test,
                                                pos_label=1, zero_division=0)
    precision, recall, _ = precision_recall_curve(y_test, mdl.predict_proba(X_test)[:,1])
    results['test_PR_AUC'] = auc(recall, precision)
```
