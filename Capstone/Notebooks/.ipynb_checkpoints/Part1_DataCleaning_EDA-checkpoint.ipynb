## 1.3 Missing Values

---

For medical related datasets missing values are typically unavoidable.  In terms of what kind of missing values this dataset has we cannot be sure. As reasons for the missing values in lab results could range across a variety of reasons. Some examples might be that a patient did not need the lab test therefore it was never requested and no value exists, or maybe a the fluid sample for a lab test was damaged and was deemed unuseable. Becasue we can not be sure if there is some sort of systematic reasoning on why some patients do not have lab results we will need to take precautionary messures in dealing with these NaN values and understnad the possible implications any one approach will have on bias in our modeling.

#### Two main approaches for dealing with missingness in clinicall data are:
1. Filtering out all records with missing data and use only the complete sets for building the models. For complete case analysis all patient records with missing data would be excluded but that would lead to a significantly smaller smaple size with reduced statistical signifigance.

2. Imputation. Prior to building any models fill the NaN values through imputation. A basic aproach could be filling NaN values with the Mean of the column values and add in missing value indicator columns that identify those values that were NaN prior to filling them with the mean. 

3. Another possible approach would be using linear regression to fill the NaN values. A Regression Imputation approach assumes that the data is Missing At Random. A major disadvantage of this method is that it could reduce the inherent variability in the imputed variable. In other words, since we substitute the missing data with regression outputs, the predicted values lie along the regression hyperplane where the variable would have actually contained some noise/bias.

The graph below is displaying the total number of patients that have been diagnosised with a mental illness vs those patients who have not been diagnosised with a mental illness. With those with a mental illness totalling just under 10,000. This indicated a major class imbalance. Most machine learning algorithms assume data equally distributed. So when we have a class imbalance, the machine learning classifier tends to be more biased towards the majority class, causing bad classification of the minority class. In order to counteract this imbalance there are couple of options. 

1. Over Sampling the minority class
2. Under sampling the majoirty class

Considering the imbalance is swayed in both cases towards the class that does not have the diagnosis the justification to use undersampling is grounded in the reasoning to not use ovesampling. An oversampling method creates sythetic or *fake* data that could cause disillusion of model accuracy when preidcting. The below graphs gave light to the imbalance and will be addressed further in the Part 2 - Baseline Modeling where several undersampling methods will be applied to pipelines in order to find the best solution.

```python
 def plot_2d_space(X, y, label='Classes'):   
    colors = ['red', '#36454f']
    markers = ['o', 'o']
    for l, c, m in zip(np.unique(y), colors, markers):
        plt.scatter(
            X[y==1, 1],
            X[y==0, 0],
            c=c, label=0, marker=m
        )
    plt.title(label)
    plt.legend(loc='upper right')
    plt.show()

#plot_2d_space(X_train_resampled, y_train_resampled, 'Tomek links under-sampling')
```

```python
from sklearn.decomposition import PCA
import pylab as pl
def plot_this(X_rs,y_rs,method):
    # Use principal component to condense the 10 features to 2 features
    pca = PCA(n_components=2).fit(X_rs)
    pca_2d = pca.transform(X_rs)
    # Assign colors
    for i in range(0, pca_2d.shape[0]):
        if y_rs[i] == 0:
            c1 = pl.scatter(pca_2d[i,0],pca_2d[i,1],c='r', marker='o')
        elif y_rs[i] == 1:c2 = pl.scatter(pca_2d[i,0],pca_2d[i,1],c='g', marker='*')
    pl.legend([c1, c2], ['Class 1', 'Class 2'])
    pl.title(method)
    pl.axis([-4, 5, -4, 4])  # x axis (-4,5), y axis (-4,4)
    pl.show()

plot_this(X_scaled,y_dem,'Original')
#view raw
```

#### 1) Tomek Links

```python
tl = TomekLinks()
X_tl, y_tl =tl.fit_resample(X_scaled, y_dem)
#tl = TomekLinks()
#tl_sampler = tl.fit_resample(X_train, y_train)
plot_2d_space(X_tl, y_tl, label='Tomek UnderSampler')
```

**Under-sampling: Tomek links**
Tomek links are pairs of very close instances, but of opposite classes. Removing the instances of the majority class of each pair increases the space between the two classes, facilitating the classification process.

One thing to remember is that these methods are not a panacea to the Class Imbalance Problem. These methods work by introducing a special kind of bias into data; one that makes it easier for predictive models to learn generalizable patterns. If you do use these methods, it’s important to understand how this special bias affects the outcomes of your model.

```python
# RandomUnderSampler
rs = RandomUnderSampler()
X_rs, y_rs = rs.fit_resample(X_scaled, y_dem)
plot_2d_space(X_rs, y_rs, label='Random UnderSampler')
```

```python
cc = ClusterCentroids()
X_cc, y_cc = cc.fit_resample(X_scaled, y_dem)
plot_2d_space(X_cc, y_cc, label='Cluster Centroid UnderSampler')
```

```python
# NeighbourhoodCleaningRule
ncr = NeighbourhoodCleaningRule()
X_NCR, y_NCR = ncr.fit_resample(X_scaled, y_dem)
plot_2d_space(X_ncr, y_ncr, label='NeighbourhoodCleaningRule UnderSampler')
```

# Principal Feature Analysis

With 59 features there are many possibilities to which could be the most influential on predicting the y-variable when modeling. Many of these featues could also be redundant and could be dropped. Below we are going to look at how all of the features variables relate to one another and summarize these relationships.
Then, use the summary and look at which combinations of the variables are most important. This can also quantify how important each combination is and rank these combinations.

Two assumptions that PCA makes:

Linearity: PCA detects and controls for linear relationships, so we assume that the data does not hold nonlinear relationships (or that we don't care about these nonlinear relationships).
We are using our covariance matrix to determine important "directions," which is a measure of the linear relationship between observations!
There are other types of feature extraction like t-SNE and PPA, though we won't formally cover those in a global lesson.
Large variances define importance: If data is spread in a direction, that direction is important! If there is little spread in a direction, that direction is not very important.

To be able to use principal Feature analysis we need to first define our X and y and train/test/split the dataset. This will be the same train/test/split done in the modeling notebook to maintain consistency. 

### Defining X and y for Mental Illness classification

```python
#defining Features for mental illness prediction 
features = df.drop(columns=['MENTAL ILLNESS','DEMENTIA','subject_id'])
#Defining X and y for mental illness classifcation 
X_ml = features
y_ml = df['MENTAL ILLNESS']
```

```python
#Train/Test/split
X_train, X_test, y_train, y_test = train_test_split(X_ml,
                                                    y_ml,
                                                    test_size = 0.20,
                                                    random_state = 7)
```

```python
X_train.shape
```

#### Scaling Data

```python
# Instantiate our StandardScaler.
ss = StandardScaler()

# Standardize X_train.
X_train = ss.fit_transform(X_train)

# Standardize X_test.
X_test = ss.transform(X_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, ExtraTreesClassifier
from sklearn.model_selection import cross_val_score, train_test_split, GridSearchCV
from sklearn.pipeline import FeatureUnion, make_pipeline
```

#### Replacing NaN values

```python
imp = SimpleImputer(missing_values=np.nan, strategy='mean')
imp.fit(X_train)
X_train_imp = imp.transform(X_train)
X_test_imp = imp.transform(X_test)
#df_imputed_mean = pd.DataFrame(df_imputed,columns=col)
```

#### Principal Conponant Analaysis

```python
# Instantiate PCA.
pca = PCA(random_state = 7)
# Fit PCA on the training data.
pca.fit(X_train_imp)
# Transform PCA on the training data.
Z_train = pca.transform(X_train_imp)
```

```python
pd.DataFrame(Z_train).describe()
```

### Number of Features to explain variance

---

Using PCA below we can estimate how many components are needed to describe the data. By looking at the cumulative explained variance ratio as a function of the number of components we can determine the number of features neeed to explain our y-variable. Below is the explained variance ratio for each component and the cumulative explained variance. 

```python
# Pull the explained variance attribute.
var_exp = pca.explained_variance_ratio_
print(f'Explained variance: {np.round(var_exp,3)}')

print('')

# Generate the cumulative explained variance.
cum_var_exp = np.cumsum(var_exp)
print(f'Cumulative explained variance: {np.round(cum_var_exp,3)}')
```

We can visualize this below to see how the cumulative explained variance increases in the number of components aka features. 

```python
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance');
```

```python
# Generate the cumulative explained variance.
cum_var_exp = np.cumsum(var_exp)
print(f'Cumulative explained variance PCA 2: {np.round(cum_var_exp[2:3],3)}')
print(f'Cumulative explained variance PCA 10: {np.round(cum_var_exp[10:11],3)}')
print(f'Cumulative explained variance PCA 15: {np.round(cum_var_exp[15:16],3)}')
print(f'Cumulative explained variance PCA 20: {np.round(cum_var_exp[20:21],3)}')
```

This curve in the graph above quantifies how much of the total, variance is contained within the first N components. Above the data shows that the first 10 components contain approximently **72%** of the variance, while approximently 20 components contain **96%** of the variance. 

This graph indicates that the two-dimensional projection loses a lot of information (as measured by the explained variance) and that we'd need about 20-25 components to retain 96%+ of the variance. Since this is a high dimenional dataset this plot means we can still retian 96% of variance with 20 components.
